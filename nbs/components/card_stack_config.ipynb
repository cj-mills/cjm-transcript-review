## Configuration

The review card stack uses a unique `sd-review` prefix to match the existing Phase 3 HTML IDs.
Cards display text content with timing and source info, using moderate width for readability.

```python
#| export
REVIEW_CS_CONFIG = CardStackConfig(
    prefix="sd-review",
    card_width_min=30,
    card_width_max=70,
    card_width_step=5,
    style=CardStackStyleConfig(
        section_gap="0.75rem",
        slot_padding="0.125rem",
        viewport_padding_x='0.0rem',
        viewport_padding_y='0.0rem'
    ),
)

REVIEW_CS_IDS = CardStackHtmlIds(prefix=REVIEW_CS_CONFIG.prefix)
REVIEW_CS_BTN_IDS = CardStackButtonIds(prefix=REVIEW_CS_CONFIG.prefix)
```
